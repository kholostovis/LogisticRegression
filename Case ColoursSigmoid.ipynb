```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
reds = list(range(127,255,2))
greens = list(range(127,255,2))


x = []
y = []
colors = []
sizes = []


for redShade in reds:
    for greenShade in greens:
        x.append(redShade)
        y.append(greenShade)
        sizes.append(50)
        colors.append((redShade, greenShade, 0))
        
colors = np.array(colors)

plt.style.use(['seaborn-dark'])
plt.scatter(x,y,c=colors/255.0, s=sizes)


borderLineX = list(range(225,260))
borderLineY = []

for i in borderLineX:
    borderLineY.append(-i+486)

    
plt.plot(borderLineX, borderLineY, 'r--')
plt.show()
```

```python
randX = np.random.randint(120,255,300)
randY = np.random.randint(120,255,300)

colors = []
sizes = []

for i in range(0,len(randX)):
    x = randX[i]
    y = randY[i]
    sizes.append(50)
    if x + y - 486 > 0:
        colors.append('yellow')
    else:
        colors.append('gray')

plt.plot(borderLineX, borderLineY, 'r--')
plt.scatter(randX, randY, c=colors, s=sizes)
plt.show()
```

```python
def sigmoid(x):
    return 1/(1 + np.math.exp(-x))


sigX = list(range(-10,10))
sigY = []

for i in sigX:
    sigY.append(sigmoid(i))

plt.plot(sigX, sigY, c='black')
plt.axvline(x=5, c='green', ls='--')
plt.axvline(x=-5, c='red', ls='--')
plt.show()


reds = list(range(200,255))
greens = list(range(200,255))

x = []
y = []
colors = []
sizes = []


for redShade in reds:
    for greenShade in greens:
        x.append(redShade)
        y.append(greenShade)
        sizes.append(400)
        probability = sigmoid(redShade+greenShade-486)
        colors.append([round(255*probability), round(255*probability), 0])

colors = np.array(colors)

plt.style.use(['seaborn-dark'])
plt.scatter(x,y, c=colors/255.0,s=sizes)
plt.show()
```
